<h1>Prompt Engineering Basics</h1>

<h2> What is prompt engineering?</h2>

 - Prompt Engineering is optimization of prompts/inputs to LLMs, to get most relevant and accurate response from LLMs
 
<h2> Why Prompt Engineering? </h2>
 
 - Even though LLMs are powerful and intelligent, they need guidance to come up with most accurate answers
 - For eg. prompts `Write a Poem` vs `Write a Poem on Motivation` have different responses
 ![title](poem.png)


<h2>Basic prompt structure</h2>
 
 -  **Instruction** - What you want the LLMs to do
 -  **Context** - Additional information needed to get the output
 -  **Input** - Input on which you want the instruction to be performed
 -  **Output** - format in which want the output

For eg: Task - Summarizing a document
```
Instruction:
Summarize the given research paper into a clear, structured, and beginner-friendly explanation. Focus on extracting the key ideas, contributions, methodology, and significance of the paper. Avoid unnecessary jargon and explain technical concepts in simple terms where possible.

Context:
The document is a highly influential machine learning research paper that introduced the Transformer architecture, which relies entirely on attention mechanisms instead of recurrent or convolutional networks. The summary should help a reader (with basic ML knowledge) quickly understand:

* Why the paper was written
* What problem it solves
* How the proposed solution works
* Why it is important today

Input document:https://proceedings.neurips.cc/paper_files/paper/2017/file/3f5ee243547dee91fbd053c1c4a845aa-Paper.pdf

Provide the output in the following structured format:
1. TL;DR (3–5 bullet points)
2. Problem Statement
3. Key Contributions
4. Core Concepts Explained
   - Self-Attention
   - Multi-Head Attention
   - Encoder-Decoder Architecture
5. Methodology (Step-by-step flow)
6. Results & Performance
7. Advantages over Previous Approaches (RNNs/CNNs)
8. Limitations (if any)
9. Why This Paper Matters (Real-world impact, modern AI relevance)
10. Simple Analogy (for intuitive understanding)
```

<h2>Prompting Techniques</h2>

1. Zero shot
2. Few Shot
3. Chain of thought prompting

<h3> Zero Shot Prompting Technique </h3>
 
Prompting with no examples, only instructions

````python
# Download ollama: https://ollama.com/download
# Download the model you want to use `ollama pull llama3`
%pip install ollama
import ollama

# print(ollama.list())
response = ollama.chat(
    model='llama3',
    messages=[
        {"role": "user", "content": "The movie was horrible. Sentiment:"}
    ]
)

print(response['message']['content'])
````

<h3>Few Shot Prompting Technique</h3>

Prompting with providing few examples of input output pairs so LLMs can learn the pattern and apply to new input

````python
response = ollama.chat(
    model='gemma3',
    messages=[
        {"role": "user", "content": """Extract person names from the text.

        Text: Elon Musk founded SpaceX.
        Names: Elon Musk

        Text: Barack Obama was the 44th president.
        Names: Barack Obama

        Text: Sundar Pichai leads Google.
        Names:"""}
    ]
)

print(response['message']['content'])
````

<h3>Chain of Thought Prompting Technique</h3>

Make the model reason step by step to come up with the answer. 
Eg. Tasks like solving a math problem

````python
response = ollama.chat(
    model='gemma3',
    messages=[
        {"role": "user", "content": """
            I have 10 apples, I gave 3 to my younger sibling, 2 to my elder sibling. 
            Then went to the market and bought 21 more apples and distributed equaly between all three of us. 
            How many apples does each one of us have?
        """}
    ]
)

print(response['message']['content'])
````

````python
response = ollama.chat(
    model='gemma3',
    messages=[
        {"role": "user", "content": """
            Question: I have 10 apples, I gave 3 to my younger sibling, 2 to my elder sibling. 
            Then went to the market and bought 21 more apples and distributed equaly between all three of us. 
            How many apples does each one of us have?
            Answer: **Step‑by‑step**

                1. **Start with the apples you originally had**  
                   - You: 10 apples  

                2. **Give apples to your siblings**  
                   - Younger sibling gets 3 → Younger now has 3  
                   - Elder sibling gets 2 → Elder now has 2  
                   - You keep the rest: \(10-3-2 = 5\) apples  
                
                   At this point:  
                
                   | Person          | Apples |
                   |-----------------|--------|
                   | You             | 5      |
                   | Younger sibling | 3      |
                   | Elder sibling   | 2      |

                3. **Buy 21 more apples**  
                
                4. **Distribute those 21 apples equally among the three of you**  
                   - \(21 \div 3 = 7\) apples each  
                
                5. **Add the newly‑received 7 apples to each person’s existing count**
                
                   | Person          | Original apples | +7 from the market | **Total** |
                   |-----------------|-----------------|-------------------|-----------|
                   | You             | 5               | 7                 | **12**    |
                   | Younger sibling | 3               | 7                 | **10**    |
                   | Elder sibling   | 2               | 7                 | **9**     |
                
                ---

                ## Answer
                - **You have 12 apples**  
                - **Your younger sibling has 10 apples**  
                - **Your elder sibling has 9 apples**  

            (These totals add up to \(12 + 10 + 9 = 31\) apples, which equals the original 10 you started with plus the 21 you bought.)

         Question: I have 100 apples, I gave 33 to my younger sibling, 22 to my elder sibling. 
            Then went to the market and bought 51 more apples and distributed equaly between all three of us. 
            How many apples does each one of us have?   
        """}
    ]
)

print(response['message']['content'])
````

````python
response = ollama.chat(
    model='gemma3',
    messages=[
        {"role": "user", "content": """
            I have 10 apples, I gave 3 to my younger sibling, 2 to my elder sibling. 
            Then went to the market and bought 21 more apples and distributed equaly between all three of us. 
            How many apples does each one of us have? 
            Lets think step by step before answering
        """}
    ]
)

print(response['message']['content'])
````

<h2> Best practices </h2>

<h4> Write <b>clear</b> and specific prompts </h4>

````python
## Write clear and specific prompts

response = ollama.chat(
    model='llama3',
    messages=[
        {"role": "user", "content": "Plan a trip to Italy"}
    ]
)

print(response['message']['content'])

response = ollama.chat(
    model='llama3',
    messages=[
        {"role": "user", "content": "Plan a family trip to Italy in month of June with a toddler of age 2.5 years for 10 days from Berlin"}
    ]
)

print(response['message']['content'])
````

<h4> Specify a <b>role</b> that AI needs to assume to answer the question effectively </h4>

````python
#role of the message can be either system, user, assistant, or tool
response = ollama.chat(
    model='gpt-oss:120b-cloud',
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant. When math is needed, respond ONLY with JSON: { \"tool\": \"calculator\", \"expression\": \"...\" }"
        },
        {
            "role": "user",
            "content": "What is 100000 multiplied by 545948309?"
        }
    ]
)
print(response['message']['content'])
````

````python
response = ollama.chat(
    model='gpt-oss:120b-cloud',
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant. When math is needed, respond ONLY with JSON: { \"tool\": \"calculator\", \"expression\": \"...\" }"
        },
        {
            "role": "user",
            "content": "What is 100000 multiplied by 545948309?"
        },
        {
            "role": "assistant",
            "content": '{ "tool": "calculator", "expression": "100000 * 545948309" }'
        },
        {"role": "tool", "content": "54594830900000"}
    ]
)
print(response['message']['content'])
````

<h4> Specify the <b>action</b> that needs to be performed by AI </h4>

````python
response = ollama.chat(
    model='gpt-oss:120b-cloud',
    messages=[
        {
            "role": "user",
            "content": """
                Action:
                Classify the sentiment of the text as Positive, Negative, or Neutral.
                
                Text:
                The service was slow but the food was good."""
        }
    ]
)
print(response['message']['content'])
````

<h4>Specify the <b>output format</b>, how the response should be structured</h4>
 

````python
response = ollama.chat(
    model='gpt-oss:120b-cloud',
    messages=[
        {
            "role": "user",
            "content": """ 
            Task:
            Extract information from the text.
            
            Action:
            Identify the person's name, profession, and city.
            
            Output format:
            Return the response as a valid JSON object with the following keys:
            - name (string)
            - profession (string)
            - city (string)
            
            Text:
            My name is Nirjhari, I am a software engineer and I live in Berlin."""
        }
    ]
)
print(response['message']['content'])
````

<h4>Specify what <b>tone/behaviour</b> the answer should use</h4>


````python
response = ollama.chat(
    model='gpt-oss:120b-cloud',
    messages=[
        {
            "role": "user",
            "content": """ 
                Task:
                Explain what is Generative AI?
                
                Behavior:
                - Use a professional and concise tone
                - Be suitable for a technical interview
                - Do not use analogies or storytelling
                
                Text:
                What is Generative AI?"""
        }
    ]
)
print(response['message']['content'])
````

<h1>Advanced Prompt Engineering Techniques</h1>

<h2> Meta Prompting </h2>

 - Meta prompting is defining structure in the prompts
 - Focuses on providing the structure the LLM should follow to reach a solution, instead of actual example like in few shot.

<h4>Advantages over few shot</h4>

 - Zero shot as no examples needed
 - Increases token efficiency
 - Minimizes influence from examples

````python
import ollama
````

````python
response = ollama.chat(
    model='llama3',
    messages=[
        {"role": "user", "content": """Problem:
            A rectangle has a length that is 4 meters more than its width. The perimeter is 40 meters. What are the dimensions?
            
            Solve the problem step by step:

                1. Let the width of the rectangle be w.
                2. Express the length in terms of w.
                3. Write the formula for perimeter of a rectangle.
                4. Substitute the values into the formula.
                5. Solve the equation.
                6. State the final dimensions clearly.
            """ 
            }
    ]
)

print(response['message']['content'])
````

<h2> Self Consistency </h2>

 - Model evaluates multiple reasoning paths to come up with the answer and then chooses the best answer
 - With few shot CoT you can improve models reasoning ability to solve more complex problems

````python

response = ollama.chat(
    model='mistral',
    messages=[
        {"role": "user", "content": """Problem:
                When I was 3 my sister was double my age. Now I’m 34 how old is my sister?
            """ 
        }
    ]
)

print(response['message']['content'])
````

<h2> Generate Knowledge Prompting </h2>

 - Models generate knowledge before answering a question.

````python

response = ollama.chat(
    model='mistral',
    messages=[
        {"role": "user", "content": """In cricket a batter can be declared out on a no ball when caught by a fielder. 
        Is this statement correct?""" 
        }
    ]
)

print(response['message']['content'])
````

````python
response = ollama.chat(
    model='mistral',
    messages=[
        {"role": "user", "content": """Input: Greece is larger than mexico.
            Knowledge: Greece is approximately 131,957 sq km, while Mexico is approximately 1,964,375 sq km, making Mexico 1,389% larger than Greece.
            Input: Glasses always fog up.
            Knowledge: Condensation occurs on eyeglass lenses when water vapor from your sweat, breath, and ambient humidity lands on a cold surface, cools, and then changes into tiny drops of liquid, forming a film that you see as fog. Your lenses will be relatively cool compared to your breath, especially when the outside air is cold.
            Input: A fish is capable of thinking.
            Knowledge: Fish are more intelligent than they appear. In many areas, such as memory, their cognitive powers match or exceed those of ’higher’ vertebrates including non-human primates. Fish’s long-term memories help them keep track of complex social relationships.
            Input: A common effect of smoking lots of cigarettes in one’s lifetime is a higher than normal chance of getting lung cancer.
            Knowledge: Those who consistently averaged less than one cigarette per day over their lifetime had nine times the risk of dying from lung cancer than never smokers. Among people who smoked between one and 10 cigarettes per day, the risk of dying from lung cancer was nearly 12 times higher than that of never smokers.
            Input: A rock is the same size as a pebble.
            Knowledge: A pebble is a clast of rock with a particle size of 4 to 64 millimetres based on the Udden-Wentworth scale of sedimentology. Pebbles are generally considered larger than granules (2 to 4 millimetres diameter) and smaller than cobbles (64 to 256 millimetres diameter).
            Input: In cricket a batter can be declared out on a no ball when caught by a fielder
            Knowledge:
            """ 
        }
    ]
)

knowledge = response['message']['content']

response = ollama.chat(
    model='mistral',
    messages=[
        {"role": "user", "content": f"""Question: In cricket a batter can be declared out on a no ball when caught by a fielder. 
                        Is this statement correct?
                        Knowledge: {knowledge}
                        Answer: 
                        """
        }
    ]
)

print(response['message']['content'])
````

<h2> Prompt Chaining </h2>

 - Breaking down a bigger and complex tasks into smaller easy to solve subtasks
 - Response from smaller subtask is given as input to another subtask to LLMs
 - This improves reliability and LLMs performance 

````python
# Prompt 1
messages = [
        {"role": "user", "content": """ 
            You are an AI assistant.

            Task:
            Create a structured outline for a beginner-friendly study guide on Machine Learning.
            
            Include:
            - Introduction
            - Core concepts
            - Simple examples
            - Practice exercises
        """ 
        }
    ]
response = ollama.chat(
    model='mistral',
    messages=messages
)

response1 = response['message']['content']

messages = [{"role": "user", "content": f""" 
                                    Using the following outline:
    
                                    {response1}
    
                                    Now expand ONLY the "Key concepts" section into a beginner-friendly explanation with examples.
                                """ }]

response = ollama.chat(
    model='mistral',
    messages=messages
)

print(response['message']['content'])
````

<h2> Tree of Thoughts Prompting </h2>

 - LLMs come up with multiple thouhts on how to solve the problem
 - Then evaluates each thought while eliminating the ones not relevant to come up with final answer
 - It uses DFS/BFS or bracktracking approaches to evaluate all the thought/options 
 - Good for tasks which require evaluating options or strategic lookahead like in solving puzzles.

 ![title](ToT.png)

```
Problem: Solve the following Sudoku puzzle.

Rules:
- Each row must contain digits 1–9 without repetition
- Each column must contain digits 1–9 without repetition
- Each 3x3 subgrid must contain digits 1–9 without repetition

Puzzle:
5 3 _ | _ 7 _ | _ _ _
6 _ _ | 1 9 5 | _ _ _
_ 9 8 | _ _ _ | _ 6 _
------+-------+------
8 _ _ | _ 6 _ | _ _ 3
4 _ _ | 8 _ 3 | _ _ 1
7 _ _ | _ 2 _ | _ _ 6
------+-------+------
_ 6 _ | _ _ _ | 2 8 _
_ _ _ | 4 1 9 | _ _ 5
_ _ _ | _ 8 _ | _ 7 9

You are an AI solving a Sudoku puzzle using Tree-of-Thought reasoning.

Step 1: Generate 3 different possible strategies to start solving the puzzle:
- Strategy A: Focus on easiest rows/columns first
- Strategy B: Focus on most constrained cells first
- Strategy C: Focus on one 3x3 grid at a time

Step 2: For each strategy:
- Apply it to make the next 3–5 moves in the puzzle
- Show intermediate board state
- Identify any contradictions or invalid placements

Step 3: Evaluate each strategy:
- Which one progresses the furthest without contradiction?
- Which one reduces uncertainty the most?

Step 4:
- Discard the strategy that leads to contradiction or least progress
- Keep the best strategy

Step 5:
- Continue expanding the best strategy step-by-step
- If a contradiction occurs, backtrack and try another branch

Step 6:
- Provide the final completed Sudoku grid
```
